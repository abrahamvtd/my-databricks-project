```python
class TableManager:
    def __init__(
        self,
        catalog_name,
        schema_name,
        table_name
    ):
        self.catalog_name = catalog_name
        self.schema_name = schema_name
        self.table_name = table_name

    def create_schema_catalog(self):
        spark.sql(
            f"CREATE CATALOG IF NOT EXISTS {self.catalog_name}"
        )
        spark.sql(
            f"CREATE SCHEMA IF NOT EXISTS {self.catalog_name}.{self.schema_name}"
        )

    def create_table(self):
        spark.sql(
            f"CREATE TABLE IF NOT EXISTS {self.catalog_name}.{self.schema_name}.{self.table_name}"
        )

    def schema_definition(self):
        spark.sql(
            f"""
            ALTER TABLE {self.catalog_name}.{self.schema_name}.{self.table_name} ADD COLUMNS (
                customer_id BIGINT,
                first_name STRING,
                last_name STRING,
                email STRING,
                country STRING,
                age BIGINT
            )
            """
        )

    def write_table(self, df):
        df.createOrReplaceTempView("tmp_view")
        spark.sql(
            f"INSERT INTO {self.catalog_name}.{self.schema_name}.{self.table_name} SELECT * FROM tmp_view"
        )

    def drop_table(self):
        spark.sql(
            f"DROP TABLE IF EXISTS {self.catalog_name}.{self.schema_name}.{self.table_name}"
        )

    def read_table(self):
        df = spark.sql(
            f"SELECT * FROM {self.catalog_name}.{self.schema_name}.{self.table_name}"
        )
        return df
```
