```python
df_silver = spark.read.table(
    "demo_catalog1.demo_schema1.demo_table1"
)
display(df_silver)
```

```python
from pyspark.sql.functions import concat_ws, upper

df_transform = (
    df_silver
    .withColumn('Fullname', concat_ws(' ', df_silver.first_name, df_silver.last_name))
    .withColumn('country', upper(df_silver.country))
    .dropDuplicates()
    .orderBy("customer_id")
)

df_transform.write.partitionBy("country").mode("overwrite").format("delta").saveAsTable("silver_table")

display(df_transform)
```
