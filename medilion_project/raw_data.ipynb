```python
import requests
import json
from pyspark.sql import SparkSession

url = "https://api.open-meteo.com/v1/forecast?latitude=40.71&longitude=-74.01&hourly=temperature_2m"

response = requests.get(url)
raw_json = response.json()

# Convert to Spark DataFrame
df_raw = spark.createDataFrame([raw_json])

# Write to Bronze (managed table)
df_raw.write.mode("overwrite").saveAsTable("medallion_project.bronze.weather_raw")
```
