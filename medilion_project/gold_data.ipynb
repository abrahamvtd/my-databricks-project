```python
from pyspark.sql.functions import to_date, avg

weather_silver = spark.table("medallion_project.silver.weather_clean")

gold_df = (
    weather_silver
    .withColumn("date", to_date("timestamp"))
    .groupBy("date")
    .agg(avg("temperature").alias("avg_daily_temperature"))
)

gold_df.write.mode("overwrite").saveAsTable("medallion_project.gold.daily_weather_summary")
```
