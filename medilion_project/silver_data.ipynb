```python
from pyspark.sql.functions import explode, col

weather_raw = spark.table("medallion_project.bronze.weather_raw")

df_exploded = (
    weather_raw
    .withColumn("hourly_data", explode(col("hourly.temperature_2m")))
    .withColumn("timestamp", explode(col("hourly.time")))
)

df_clean = df_exploded.select(
    col("timestamp"),
    col("hourly_data").alias("temperature")
)

df_clean.write.mode("overwrite").saveAsTable("medallion_project.silver.weather_clean")
```
